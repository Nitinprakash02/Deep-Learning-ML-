```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y =make_blobs(random_state=1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.scatter(X[y == 1, 0], X[y == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y == 2, 0], X[y == 2, 1], s = 100, c = 'red', label = 'Cluster 3')
plt.title('Clusters of Isotropic Gaussian blobs')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
plt.scatter(X[kmeans.labels_ == 0, 0], X[kmeans.labels_ == 0, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.scatter(X[kmeans.labels_ == 1, 0], X[kmeans.labels_ == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[kmeans.labels_ == 2, 0], X[kmeans.labels_ == 2, 1], s = 100, c = 'red', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 200, c = 'yellow', label = 'Centroids')
plt.title('KMeans Clustering Result - 3 Clusters')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()
```

```python
kmeans=KMeans(n_clusters=5)
kmeans.fit(X)

plt.scatter(X[kmeans.labels_ == 0, 0], X[kmeans.labels_ == 0, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.scatter(X[kmeans.labels_ == 1, 0], X[kmeans.labels_ == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[kmeans.labels_ == 2, 0], X[kmeans.labels_ == 2, 1], s = 100, c = 'red', label = 'Cluster 3')
plt.scatter(X[kmeans.labels_ == 3, 0], X[kmeans.labels_ == 3, 1], s = 100, c = 'purple', label = 'Cluster 4')
plt.scatter(X[kmeans.labels_ == 4, 0], X[kmeans.labels_ == 4, 1], s = 100, c = 'orange', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 200, c = 'yellow', label = 'Centroids')
plt.title('KMeans Clustering Result - 5 Clusters')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()
```

## **DBSCAN**

```python
X, y = make_moons(n_samples=200, noise=0.05, random_state= 0)
```

**Orignal Data**

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.scatter(X[y == 1, 0], X[y == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.title('Clusters two interleaving half circle')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN() # Default setting, eps=0.5, min_samples=5
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
plt.scatter(X[dbscan.labels_ == 0, 0], X[dbscan.labels_ == 0, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.title('KMeans Clustering Result - 1 Cluster')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()
```

**Scaling the data**

Finding a good setting for epsilon is sometimes easier after scaling the data using StandardScaler, as using these scaling techniques will ensure that all features have similar ranges.

```python
from sklearn.preprocessing import StandardScaler

scaler =StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
plt.scatter(X_scaled[y == 0, 0], X_scaled[y == 0, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.scatter(X_scaled[y == 1, 0], X_scaled[y == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.title('Clusters two interleaving half circle')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN()
dbscan.fit(X_scaled)
```

```python
plt.scatter(X_scaled[dbscan.labels_ == 0, 0], X_scaled[dbscan.labels_ == 0, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.scatter(X_scaled[dbscan.labels_ == 1, 0], X_scaled[dbscan.labels_ == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.title('KMeans Clustering Result - 2 Clusters')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()
```
