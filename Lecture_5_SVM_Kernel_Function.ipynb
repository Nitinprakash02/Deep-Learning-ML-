```python
X= dataset [['x1', 'x2']]
y= dataset['ring']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train.shape, X_test.shape
```

```python
# import SVC classifier
from sklearn.svm import SVC

# import metrics to compute accuracy
from sklearn.metrics import accuracy_score

linear_svc=SVC(kernel='linear') 

# fit classifier to training set
linear_svc.fit(X_train,y_train)


# make predictions on test set
y_pred_test=linear_svc.predict(X_test)
```

```python
y_pred_test
```

```python
y_test
```

```python
# compute and print accuracy score
print('Model accuracy score with linear kernel : {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```

```python
linear_svc=SVC() # default='rbf'

# fit classifier to training set
linear_svc.fit(X_train,y_train)


# make predictions on test set
y_pred_test=linear_svc.predict(X_test)


# compute and print accuracy score
print('Model accuracy score with linear kernel : {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```

```python
linear_svc=SVC(kernel='poly', degree=2) 

# fit classifier to training set
linear_svc.fit(X_train,y_train)


# make predictions on test set
y_pred_test=linear_svc.predict(X_test)


# compute and print accuracy score
print('Model accuracy score with Polynomial kernel: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```

```python
linear_svc=SVC(kernel='poly', degree=3) #DOUBLE CHECK DEGREE=3 in RAPIDMINDER

# fit classifier to training set
linear_svc.fit(X_train,y_train)


# make predictions on test set
y_pred_test=linear_svc.predict(X_test)


# compute and print accuracy score
print('Model accuracy score with Polynomial kernel: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```
